```python
import pandas as pd
import risk_kit as rk
import numpy as np
%load_ext autoreload
%autoreload 2

ind_return = rk.get_ind_returns()
tmi_return = rk.get_total_market_index_returns()
```

```python
risky_r = ind_return["2000":][["Steel", "Fin", "Beer"]]
# Safe Asset
safe_r = pd.DataFrame().reindex_like(risky_r)
```

```python
safe_r[:] = 0.03/12
start = 1000
floor = 0.80
```

1.  Cushion - (Asset Value minus Floor Value)
2. Compute an Allocation to Safe and Risky Assets --> m*risk budget
3. Recompute the Asset Value bases on the returns

```python
dates = risky_r.index
n_steps = len(dates)
account_value = start
floor_value = start*floor
m = 3
account_history = pd.DataFrame().reindex_like(risky_r)
cushion_history = pd.DataFrame().reindex_like(risky_r)
risky_w_history = pd.DataFrame().reindex_like(risky_r)

for step in range(n_steps):
    cushion = (account_value - floor_value)/account_value
    risky_w = m*cushion
    risky_w = np.minimum(risky_w,1)
    risky_w = np.maximum(risky_w,0)
    safe_w = 1-risky_w
    risky_allocation = account_value*risky_w
    safe_allocation = account_value*safe_w
    ## Update the account value for this time step
    account_value = risky_allocation*(1+risky_r.iloc[step]) + safe_allocation*(1+safe_r.iloc[step])
    # Save values to plot history
    cushion_history.iloc[step] = cushion
    risky_w_history.iloc[step] = risky_w
    account_history.iloc[step] = account_value
    
    
    
```

```python
risky_wealth = start*(1+risky_r).cumprod()
```
